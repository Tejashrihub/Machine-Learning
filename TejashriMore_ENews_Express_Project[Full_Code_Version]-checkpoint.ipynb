### Univariate Analysis

```python
# function to plot a boxplot and a histogram along the same scale.

def histogram_boxplot(data, feature, figsize=(12, 7), kde=False, bins=None):
   
    f2, (ax_box2, ax_hist2) = plt.subplots(
        nrows=2,  # Number of rows of the subplot grid= 2
        sharex=True,  # x-axis will be shared among all subplots
        gridspec_kw={"height_ratios": (0.25, 0.75)},
        figsize=figsize,
    )  # creating the 2 subplots
    sns.boxplot(
        data=data, x=feature, ax=ax_box2, showmeans=True, color="gray"
    )  # boxplot will be created and a star will indicate the mean value of the column
    sns.histplot(
        data=data, x=feature, kde=kde, ax=ax_hist2, bins=bins, palette="Spectral"
    ) if bins else sns.histplot(
        data=data, x=feature, kde=kde, ax=ax_hist2, 
    )  # For histogram
    ax_hist2.axvline(
        data[feature].mean(), color="red", linestyle="--"
    )  # Add mean to the histogram
    ax_hist2.axvline(
        data[feature].median(), color="pink", linestyle="-"
    )  # Add median to the histogram
```

```python
# function to create labeled barplots

def labeled_barplot(data, feature, perc=False, n=None):
    
    total = len(data[feature])  # length of the column
    count = data[feature].nunique()
    if n is None:
        plt.figure(figsize=(count + 1, 5))
    else:
        plt.figure(figsize=(n + 1, 5))

    plt.xticks(rotation=0, fontsize=12)
    ax = sns.countplot(data=data, x=feature, palette="Spectral", order=data[feature].value_counts().index[:n].sort_values())

    for p in ax.patches:
        if perc == True:
            label = "{:.1f}%".format(100 * p.get_height() / total)  # percentage of each class of the category
        else:
            label = p.get_height()  # count of each level of the category

        x = p.get_x() + p.get_width() / 2  # width of the plot
        y = p.get_height()                 # height of the plot

        ax.annotate(label, (x, y), ha="center", va="center", size=12, xytext=(0, 5), textcoords="offset points")  # annotate the percentage

    plt.show()  # show the plot
```

## 2. Do the users spend more time on the new landing page than the existing landing page? (10 Marks)

### Perform Visual Analysis

```python
# visual analysis of the time spent on the new page and the time spent on the old page
plt.figure(figsize=(10,5))
sns.boxplot(x = 'landing_page', y = 'time_spent_on_the_page', data = df,palette = 'Spectral')
plt.show()
```

#### Observations:
* The box plot further confirms our statistical finding that the users spend more time on the new landing page than they do on the old one.

### Step 1: Define the null and alternate hypotheses


#### Null Hypothesis: Users spend equal time on the old and new landing page.

#### Alternate Hypothesis: Users spend greater amount of time on the new landing page.

#### Let  ðœ‡1,ðœ‡2  be the mean time spent on the new and old landing page respectively.

#### ð»0:ðœ‡1=ðœ‡2

#### ð»ð‘Ž:ðœ‡1>ðœ‡2


### Step 2: Select Appropriate test

#### Continuous Data
#### Independent Populations
#### Random Sampling from the population
#### Test for equality of two means 
#### Unknown standard deviation
### Hence, We can use two sample independent T-test for this problem.


### Step 3: Decide the significance level

#### Consider a significance level 0.05

### Step 4: Collect and prepare data

```python
# creating subsetted data frame for new landing page users 
time_new = df[df['landing_page'] == 'new']['time_spent_on_the_page']
# creating subsetted data frame for old landing page users
time_old= df[df['landing_page'] == 'old']['time_spent_on_the_page']
# find the sample means and sample standard deviations for the two samples
df.groupby('landing_page').describe()
```

#### Observations:
* The sample standard deviation of the time spent on the new page and the old page is 1.81 and 2.58 respectively.


### Step 5: Calculate the p-value

```python
# import the required functions
from scipy.stats import ttest_ind
# find the p-value
test_stat, p_value = ttest_ind(time_new, time_old, equal_var = False, alternative = 'greater')
print('The p-value is ', p_value)
```

#### Observations:
* The p-value is 0.0001392 which is less than significance level.

### Step 6: Compare the p-value with $\alpha$

```python
# print the conclusion based on p-value
if p_value < 0.05:
    print(f'As the p-value {p_value} is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'As the p-value {p_value} is greater than the level of significance, we fail to reject the null hypothesis.')
```

#### Observations:
* The p-value 0.0001392381225166549 is less than the level of significance, we reject the null hypothesis.

### Step 7:  Draw inference

#### Since the p-value is less than the 5% significance level, we reject the null hypothesis. Hence,we do not have enough evidence to support the claim that the users spend the same amount of time on the old landing page as they do on the new landing page.
#### The users spend more time on the new landing page.


## 3. Is the conversion rate (the proportion of users who visit the landing page and get converted) for the new page greater than the conversion rate for the old page? (10 Marks)

### Perform Visual Analysis

```python
plt.figure(figsize=(10,5))
sns.countplot(x ='landing_page',hue='converted',palette="Pastel2", data = df)
plt.show()
```

#### Observations:
* The bar plot further confirms our statistical finding that the conversion rate is higher for users from the new landing page

### Step 1: Define the null and alternate hypotheses

#### Null Hypothesis: The conversion rate for the new page is equal to that of the old page.
#### Alternate Hypothesis:The conversion rate for the new page is greater than that of the old page.
#### Let p1, p2 be the rate of conversion for the new and old landing page respectively.


#### ð»0:p1=p2

#### ð»ð‘Ž:p1>p2


### Step 2: Select Appropriate test

#### Independent Populations
#### Random Sampling from the populations
#### Test for equality of two population proportions
### Hence, We can use two proportions Z-test for this problem.

### Step 3: Decide the significance level

#### Consider a significance level 0.05

### Step 4: Collect and prepare data

```python
# calculate the number of converted users in the treatment group
new_converted = df[df['group'] == 'treatment']['converted'].value_counts()['yes']
print('The no of converted users in the treatment group:',new_converted)

# calculate the number of converted users in the control group
old_converted = df[df['group'] == 'control']['converted'].value_counts()['yes']
print('The no of converted users in the control group:',old_converted)

n_control = df.group.value_counts()['control'] # total number of users in the control group
n_treatment = df.group.value_counts()['treatment'] # total number of users in the treatment group
print('The numbers of users served the new and old pages are {0} and {1} respectively'.format(n_control, n_treatment ))
```

#### Observations:
* In the DataFrame users who converted to subscriber from treatment group are 33 and from control group are 21.

### Step 5: Calculate the p-value

```python
# import the required fuction
from statsmodels.stats.proportion import proportions_ztest
# set the counts of conversions from new and old pages.
conversion_count = np.array([33, 21])
# set the sample sizes
group_count = np.array([50, 50])
# find the p-value
test_stat, p_value = proportions_ztest(conversion_count, group_count, alternative = 'larger')
print('The p-value is ' + str(p_value))
```

#### Observations:
* The p-value is 0.0080 which is less than significance level.

### Step 6: Compare the p-value with $\alpha$

```python
# print the conclusion based on p-value
if p_value < 0.05:
    print(f'As the p-value {p_value} is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'As the p-value {p_value} is greater than the level of significance, we fail to reject the null hypothesis.')
```

#### Observations:
* The p-value 0.0080 is less than the level of significance, we reject the null hypothesis.

### Step 7:  Draw inference

#### Since the p-value is less than the 5% significance level, we reject the null hypothesis. Hence, we do not have enough evidence to support the claim that the conversion rate for the new page is the same as the conversion rate for the old page.
#### The conversion rate for the new page is greater than the conversion rate for the old page.

#### Observations:

### Step 1: Define the null and alternate hypotheses

#### Null Hypothesis(ð»0): The converted status is independent of the preferred language.
#### Alternate Hypothesis(ð»a): The converted status depends on the preferred language

### Step 2: Select Appropriate test

#### Random sampling from the population
#### Categorical variables
### Hence, We can use Chi Square test of independence for this problem.

### Step 3: Decide the significance level

#### Consider a significance level 0.05

### Step 4: Collect and prepare data

```python
#creating a contingency table showing the distribution of the two categorical variables
contingency_table=pd.crosstab(df['converted'],df['language_preferred'])
contingency_table                                               
```

#### Observations:
* In the DataFrame 21,15,18 users gets converted as subscriber by prefering language english,french and spanish respectively. 

### Step 5: Calculate the p-value

```python
# import the required function
from scipy.stats import chi2_contingency
# find the p-value
chi, p_value, dof, expected = chi2_contingency(contingency_table)
print('The p-value is', p_value)
```

#### Observations:
* The p-value is 0.21 which is greater than significance level.

### Step 6: Compare the p-value with $\alpha$

```python
# print the conclusion based on p-value
if p_value < 0.05:
    print(f'As the p-value {p_value} is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'As the p-value {p_value} is greater than the level of significance, we fail to reject the null hypothesis.')
```

#### Observations:
* The p-value 0.2129 is greater than the level of significance, we fail to reject the null hypothesis.

### Step 7: Draw inference

#### Since the p-value is greater than the 5% significance level, we cannot reject the null hypothesis. Hence, have enough evidence to support the claim that the conversion rate does not depend on the preferred language.
#### The converted status is independent of the preferred language.

## 5. Is the time spent on the new page same for the different language users? (10 Marks)

### Perform Visual Analysis

```python
# visually plot the time spent on the new page for different language users
data_new = df[(df['landing_page']=="new")]
plt.figure(figsize=(10,5))
sns.boxplot(data = data_new , x='language_preferred',y='time_spent_on_the_page',palette = 'GnBu');
plt.xticks(rotation = 0)
plt.show()
```

#### Observations:

```python
# mean of time spent on new page by different language users
data_new = df[(df['landing_page']=="new")]
print(data_new.groupby("language_preferred")["time_spent_on_the_page"].mean())
```

#### Observations:
* In the DataFrame the mean time spent on new page by different language users are 6.6,6.1,5.8 for english,french and spanish respectively.

### Step 1: Define the null and alternate hypotheses

#### Null Hypothesis: The mean time spent on the new page by all language users is equal.
#### Alternate Hypothesis:At least one mean is unequal.
#### Let  ðœ‡1,ðœ‡2,ðœ‡3 be the means of time spent on the new page by English, French and Spanish users respectively.

#### ð»0:ðœ‡1=ðœ‡2=ðœ‡3

#### ð»ð‘Ž:  At least one mean is different from the rest.


### Step 2: Select Appropriate test

#### Test for means for more than two populations
#### The Populations are normally distributed
### Hence, We can use One-way-ANOVA test for this problem.

* For testing of normality, Shapiro-Wilkâ€™s test is applied to the response variable.

* For equality of variance, Levene test is applied to the response variable.

### Shapiro-Wilkâ€™s test

#### ð»0: Time spent on the new page follows a normal distribution


#### ð»ð‘Ž:Time spent on the new page does not follow a normal distribution

```python
# Assumption 1: Normality
# import the required function
from scipy.stats import shapiro

# find the p-value
w, p_value = shapiro(data_new['time_spent_on_the_page']) 
print('The p-value is', p_value)
```

#### Observations:
* Since p-value of the test is very large than the 5% significance level, we fail to reject the null hypothesis that the response follows the normal distribution.

### Leveneâ€™s test


#### ð»0: All the population variances are equal


#### ð»ð‘Ž: At least one variance is different from the rest

```python
#Assumption 2: Homogeneity of Variance
#import the required function
from scipy.stats import levene
statistic, p_value = levene(data_new[data_new['language_preferred']=="English"]['time_spent_on_the_page'], 
                            data_new[data_new['language_preferred']=="French"]['time_spent_on_the_page'], 
                            data_new[data_new['language_preferred']=="Spanish"]['time_spent_on_the_page'])
# find the p-value
print('The p-value is', p_value)
```

#### Observations:
* Since the p-value is large than the 5% significance level, we fail to reject the null hypothesis of homogeneity of variances.

### Step 3: Decide the significance level

#### Consider a significance level 0.05

### Step 5: Calculate the p-value

```python
from scipy.stats import  f_oneway

statistic, p_value =  f_oneway( data_new[data_new['language_preferred']=="English"]['time_spent_on_the_page'], 
                             data_new[data_new['language_preferred']=="French"]['time_spent_on_the_page'], 
                             data_new[data_new['language_preferred']=="Spanish"]['time_spent_on_the_page'])
# find the p-value
print('The p-value is', p_value)
```

#### Observations:
* The p-value is 0.43 which is greater than significance level.

### Step 6: Compare the p-value with $\alpha$

```python
# print the conclusion based on p-value
if p_value < 0.05:
    print(f'As the p-value {p_value} is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'As the p-value {p_value} is greater than the level of significance, we fail to reject the null hypothesis.')
```

#### Observations:
* The p-value 0.43 is greater than the level of significance, we fail to reject the null hypothesis.

### Step 7:  Draw inference

#### Since the p-value is greater than the 5% significance level, we cannot reject the null hypothesis. Hence, have enough evidence to support the claim that the mean time spent on the new page is equal across the different language users.
#### The mean time spent on the new page same for the different language users.
